## 2. Download NLTK resources

```python
import nltk
for pkg in ['punkt','punkt_tab','stopwords','wordnet','omw-1.4']:
    nltk.download(pkg, quiet=True)
print('NLTK ready')
```

## 4. Load the chatbot engine

```python
from utils.chatbot_engine import FAQChatbot
bot = FAQChatbot('dataset/faq.csv', threshold=0.30, log_path='logs/chat_logs.csv')
print(f'Loaded {len(bot.questions)} FAQs')
```

## 5. Quick tests

```python
for q in ['What is Machine Learning?', 'who created python', 'asdkjasd random text', '']:
    r = bot.ask(q)
    print(f'Q: {q!r}\n  -> {r.answer}\n  conf={r.confidence:.2f}\n')
```
